# **Standardization of Nmerical variable**

Standardization (Z-score Normalization)

**Concept:**
Standardization, also known as Z-score normalization, is a technique used to scale the features of a dataset so that they have a mean of 0 and a standard deviation of 1. This is particularly useful when the features of the dataset have different units or scales, as it helps to bring them to a common scale without distorting the differences in the ranges of the values.

**Formula:**
The Z-score normalization formula is:

𝑧 = (𝑥 - 𝜇) / 𝜎


**Where:**

𝑥 : x is the original value.

𝜇 : μ is the mean of the feature.

𝜎 : σ is the standard deviation of the feature.

𝑧 : z is the standardized value.

```python
standardized_df = df.copy()

for col in numerical_columns:
    mean = df[col].mean()
    std = df[col].std()
    standardized_df[col] = (df[col] - mean) / std
```

```python
plt.figure(figsize=(15, 20))

# Before standardization
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(6, 2, 2*i-1)
    plt.hist(df[col])
    plt.title(f'Before Standardization: {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

# After standardization
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(6, 2, 2*i)
    plt.hist(standardized_df[col])
    plt.title(f'After Standardization: {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

# **Coversion of categorical data into dummy variables**


One-hot encoding is a method used to convert categorical variables into a format that can be provided to ML algorithms to improve predictions. In one-hot encoding, each unique category value is converted into a new binary (0 or 1) column.

# The advantages of using one hot encoding include:

It allows the use of categorical variables in models that require numerical input.

It can improve model performance by providing more information to the model about the categorical variable.

It can help to avoid the problem of ordinality, which can occur when a categorical variable has a natural ordering (e.g. “small”, “medium”, “large”).

```python
encoded_df = pd.get_dummies(df, columns=['Day', 'SKU', 'City', 'BU', 'Brand', 'Model'], dtype=int)
encoded_df.head(10)
```
